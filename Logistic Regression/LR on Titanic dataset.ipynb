```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
train_data=pd.read_csv("0000000000002429_training_titanic_x_y_train.csv",delimiter=",")
test_data=pd.read_csv("0000000000002429_test_titanic_x_test.csv",delimiter=",")
train_data.shape,test_data.shape
```

```python
del train_data["Name"]
del test_data["Name"]
del train_data["Ticket"]
del test_data["Ticket"]
train_data.head(25)
```

```python
def f1(s):
    if s=="S":
        return 1
    else:
        return 0
#
def f2(s):
    if s=="Q":
        return 1
    else:
        return 0
#
def f3(s):
    if s=="C":
        return 1
    else:
        return 0
```

```python
train_data["S"]=train_data.Embarked.apply(f1)
test_data["S"]=test_data.Embarked.apply(f1)
train_data["Q"]=train_data.Embarked.apply(f2)
test_data["Q"]=test_data.Embarked.apply(f2)
train_data["C"]=train_data.Embarked.apply(f3)
test_data["C"]=test_data.Embarked.apply(f3)
del train_data["Embarked"]
del test_data["Embarked"]
train_data
```

```python
import numpy as np

def f4(x):
    if x==np.nan:
        return 0
    else:
        return 1
#
def f5(x):
    if x=="male":
        return 1
    else:
        return 0
```

```python
train_data["Cabin"]=train_data.Cabin.apply(f4)
test_data["Cabin"]=test_data.Cabin.apply(f4)
train_data["Sex"]=train_data.Sex.apply(f5)
test_data["Sex"]=test_data.Sex.apply(f5)
train_data
```

```python
train_data["Age"].fillna(train_data.Age.mean(),inplace=True)
test_data["Age"].fillna(test_data.Age.mean(),inplace=True)
train_data.describe()
```

```python
y_train=train_data["Survived"]
y_train=y_train.values
y_train
x_train=train_data.drop("Survived",axis=1)
x_train=x_train.values
x_test=test_data.values
x_train
```

```python
clf=LogisticRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
y_test_pred=clf.predict(x_test)
```

```python
clf.score(x_test,y_test_pred)
```

```python
np.savetxt('titanic_pred.csv',y_test_pred)
```

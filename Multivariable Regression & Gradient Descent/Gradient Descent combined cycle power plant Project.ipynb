```python
import numpy as np
```

```python
train_data=np.loadtxt("0000000000002419_training_ccpp_x_y_train.csv",delimiter=",")
test_data=np.loadtxt("0000000000002419_test_ccpp_x_test.csv",delimiter=",")
train_data.shape, test_data.shape
```

```python
y_train=train_data[:,4]
x_train=train_data[:,0:4]
```

```python
# x_train=np.column_stack((x_train,ones_train))
# x_test=np.column_stack((test_data,ones_test))
x_train=np.insert(x_train,4,1,axis=1)
x_test=np.insert(test_data,4,1,axis=1)
x_train.shape, x_test.shape
```

```python
def step_gradient(x,y,learning_rate,m):
    m_slope=np.zeros(x.shape[1])
    N=len(x)
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            m_slope[j]+=(-2/N)*(y[i]-(m.dot(x[i,:])))*x[i,j]
            m[j]=m[j]-learning_rate*m_slope[j]
    return m
```

```python
def cost(x,y,m):
    total_cost=0
    N=x.shape[0]
    for i in range(N):
        total_cost+=(1/N)*((y[i]-m.dot(x[i,:]))**2)
    return total_cost
```

```python
def gd(x,y,learning_rate,num_iterations):
    m=np.zeros(x.shape[1])
    for i in range(num_iterations):
        m=step_gradient(x,y,learning_rate,m)
        print(i,"cost :",cost(x,y,m))
    return m
```

```python
 def run():
        learning_rate=0.0000005
        num_iterations=130
        m=gd(x_train,y_train,learning_rate,num_iterations)
        return m
```

```python
m=run()
```

```python
# predictions
y_pred=np.zeros(x_test.shape[0])
for i in range(x_test.shape[0]):
    y_pred[i]=m.dot(x_test[i,:])
```

```python
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
```

```python
train_data=np.loadtxt("0000000000002419_training_ccpp_x_y_train.csv",delimiter=",")
test_data=np.loadtxt("0000000000002419_test_ccpp_x_test.csv",delimiter=",")
train_data.shape
```

```python
y_train=train_data[:,4]
x_train=train_data[:,0:4]
```

```python
x_train.shape, test_data.shape
```

```python
alg=GradientBoostingRegressor()
```

```python
alg.fit(x_train,y_train)
```

```python
test_pred=alg.predict(test_data)
```

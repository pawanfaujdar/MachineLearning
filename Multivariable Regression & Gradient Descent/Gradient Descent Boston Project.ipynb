```python
import numpy as np
```

```python
train_data=np.loadtxt("0000000000002417_training_boston_x_y_train.csv",delimiter=",")
test_data=np.loadtxt("0000000000002417_test_boston_x_test.csv",delimiter=",")
```

```python
one_train=np.ones(len(train_data))
one_test=np.ones(len(test_data))
train_data.shape,test_data.shape
```

```python
y_train=train_data[:,13]
y_train.shape
```

```python
x_train=train_data[:,0:13]
x_train.shape
```

```python
x_train=np.column_stack((x_train,one_train))
x_test=np.column_stack((test_data,one_test))
x_train.shape,x_test.shape
```

```python
def step_gradient(x,y,learning_rate,m):
    m_slope=np.zeros(x.shape[1])
    N=len(x)
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            m_slope[j]+=(-2/N)*(y[i]-(m.dot(x[i,:])))*x[i,j]
            m[j]=m[j]-learning_rate*m_slope[j]
#     new_m=m-learning_rate*m_slope 
    return m
```

```python
def cost(x,y,m):
    total_cost=0
    N=x.shape[0]
    for i in range(N):
        total_cost+=(1/N)*((y[i]-m.dot(x[i,:]))**2)
    return total_cost
```

```python
def gd(x,y,learning_rate,num_iterations):
    m=np.zeros(x.shape[1])
    for i in range(num_iterations):
        m=step_gradient(x,y,learning_rate,m)
        print(i,"cost:",cost(x,y,m))
    return m
```

```python
def run():
    learning_rate=0.003
    num_iterations=80
    m=gd(x_train,y_train,learning_rate,num_iterations)
    return m
```

```python
m=run()
```

```python
# Predictions
y_pred=np.zeros(x_test.shape[0])
for i in range(x_test.shape[0]):
    y_pred[i]=m.dot(x_test[i,:])
```

```python
np.savetxt("boston_pred.csv",y_pred)
```

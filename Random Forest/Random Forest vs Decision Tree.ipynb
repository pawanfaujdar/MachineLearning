```python
import numpy as np
import pandas as pd
```

```python
titanic_train=pd.read_csv("0000000000002429_training_titanic_x_y_train.csv",delimiter=",")
x_test=pd.read_csv("0000000000002429_test_titanic_x_test.csv",delimiter=",")
```

```python
def f(x):
    if x=="male":
        return 1
    else:
        return 0
def f1(x):
    if x=="S":
        return 1
    else:
        return 0
def f2(x):
    if x=="Q":
        return 1
    else:
        return 0
def f3(x):
    if x=="C":
        return 1
    else:
        return 0
```

```python
del titanic_train["Name"]
del x_test["Name"]
del titanic_train["Ticket"]
del x_test["Ticket"]
del titanic_train["Cabin"]
del x_test["Cabin"]
del titanic_train["Fare"]
del x_test["Fare"]
```

```python
titanic_train=titanic_train.fillna(titanic_train.median())
x_test=x_test.fillna(x_test.median())
titanic_train.dropna(inplace=True)
x_test.dropna(inplace=True)
titanic_train.isnull().sum()
```

```python
x_train=titanic_train.iloc[:,0:titanic_train.shape[1]-1]
y_train=titanic_train.iloc[:,-1]
```

```python
x_train["Sex"]=x_train.Sex.apply(f)
x_test["Sex"]=x_test.Sex.apply(f)
###########################################
x_train["S"]=x_train.Embarked.apply(f1)
x_test["S"]=x_test.Embarked.apply(f1)
x_train["Q"]=x_train.Embarked.apply(f2)
x_test["Q"]=x_test.Embarked.apply(f2)
x_train["C"]=x_train.Embarked.apply(f3)
x_test["C"]=x_test.Embarked.apply(f3)
x_train.drop("Embarked",axis=1,inplace=True)
x_test.drop("Embarked",axis=1,inplace=True)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_pred)
```

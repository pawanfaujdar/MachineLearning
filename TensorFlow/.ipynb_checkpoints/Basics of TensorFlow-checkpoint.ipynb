# Constants:

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
a=tf.constant(2)
b=tf.constant(3)
b
```

```python
c=a+b
c
```

```python
sess=tf.Session()
sess.run(c)

###################   OR     #################

sess=tf.Session()
a_val=sess.run(c)
a_val
```

```python
a1=tf.constant([[3,3]])
a2=tf.constant([[3],[3]])
```

```python
res=tf.matmul(a1,a2)
sess.run(res)
```

# Session:

```python
a=tf.constant(2)
b=tf.constant(3)
c=a+b  #or tf.add(a,b)
```

```python
with tf.Session() as sess:
    print(sess.run(c))
    
###############  OR  ################

with tf.Session() as sess:
    print(c.eval())
```

# Variables:

```python
var1=tf.Variable(100)
var1
```

```python
var2=tf.Variable(3)
sum=tf.add(var1,var2)
sess=tf.Session()
sess.run(tf.global_variables_initializer()) #without this line, we will get Error
sess.run(sum)
```

```python
var1.assign(1232)
print(sess.run(var1))
```

```python
sess.run(var1.assign(1232))
print(sess.run(var1))
```

```python
assign=var1.assign(1232)
sess.run(assign)
sess.run(var1)
```

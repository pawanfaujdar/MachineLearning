```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
import pandas as pd
import numpy as np
```

```python
df_train=pd.read_csv("mnist/train.csv")
df_test=pd.read_csv("mnist/test.csv")
df_train.shape
```

```python
x_train=df_train.iloc[:,1:]
x_test=df_test.copy()
x_train
```

```python
y_train=df_train.iloc[:,0]
y_train
```

```python
x_train=x_train.values
x_test=x_test.values
x_train.shape
```

```python
y_train=tf.one_hot(y_train,depth=10)
```

```python
sess=tf.Session()
y_train=sess.run(y_train)
y_train.shape
```

```python
import matplotlib.pyplot as plt
```

```python
first_image=x_train[1]
first_image=first_image.reshape((28,28))
plt.imshow(first_image)
plt.show()
```

```python
n_input=x_train.shape[1]
n_hidden_1=256
n_hidden_2=256
n_classes=10
```

# Initialising weights and biases:

```python
weights={
    "h1":tf.Variable(tf.random_normal([n_input,n_hidden_1])),
    "h2":tf.Variable(tf.random_normal([n_hidden_1,n_hidden_2])),
    "out":tf.Variable(tf.random_normal([n_hidden_2,n_classes]))
}
```

```python
biases={
    "h1":tf.Variable(tf.random_normal([n_hidden_1])),
    "h2":tf.Variable(tf.random_normal([n_hidden_2])),
    "out":tf.Variable(tf.random_normal([n_classes]))
}
```

# Forward Propagation:

```python
def forward_propagation(x,weights,bias):
    in_layer1=tf.add(tf.matmul(x,weights["h1"]),biases["h1"])
    out_layer1=tf.nn.relu(in_layer1)
    
    in_layer2=tf.add(tf.matmul(out_layer1,weights["h2"]),biases["h2"])
    out_layer2=tf.nn.relu(in_layer2)
    
    output=tf.add(tf.matmul(out_layer2,weights["out"]),biases["out"])
    return output
```

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())
x=tf.placeholder("float",[None,n_input])
y=tf.placeholder(tf.int32,[None,n_classes])
```

```python
pred=forward_propagation(x,weights,biases)
predictions=tf.argmax(pred,1)
true_labels=tf.argmax(y,1)
correct_predictions=tf.equal(predictions,true_labels)
```

```python
predictions_eval,labels,correct_pred=sess.run([predictions,true_labels,correct_predictions],feed_dict={x:x_train,y:y_train})
predictions_eval,labels,correct_pred
```

```python
correct_pred.sum()  #11.2%
```

# Cost:

```python
cost=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=pred,labels=y))
```

# Optimizer:

```python
optimizer=tf.train.AdamOptimizer(learning_rate=0.01)
optimize=optimizer.minimize(cost)
```

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())
```

# Batch Gradient Descent:

```python
batch_size=100
for i in range(25):
    num_batches=int(x_train.shape[0]/batch_size)
    total_cost=0
    for j in range(num_batches):
        batch_x,batch_y=x_train[j*batch_size:(j+1)*batch_size],y_train[j*batch_size:(j+1)*batch_size]
        c,_=sess.run([cost,optimize],feed_dict={x:batch_x,y:batch_y})
        total_cost+=c
    print(total_cost)
```

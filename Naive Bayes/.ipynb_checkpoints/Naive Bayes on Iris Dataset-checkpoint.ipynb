```python
def fit(X_train,Y_train):
    result={}
    class_values=set(Y_train)
    for current_class in class_values:
        result[current_class]={}
        result["total_data"]=len(Y_train)
        current_class_rows=(Y_train==current_class)
        X_train_current=X_train[current_class_rows]
        Y_train_current=Y_train[current_class_rows]
        result[current_class]["total_count"]=len(Y_train_current)
        num_features=X_train.shape[1]
        for j in range(1,num_features+1):
            result[current_class][j]={}
            all_possible_values=set(X_train[:,j-1])
            for current_value in all_possible_values:
                result[current_class][j][current_value]=(X_train_current[:,j-1]==current_value).sum()
    return result
```

```python
def Probability(dictionary,x,current_class):
#     output = np.log(dictionary[current_class]["total_count"]) - np.log(dictionary["total_data"])
    output=dictionary[current_class]["total_count"]/dictionary["total_data"]
#     output=np.log(dictionary[current_class]["total_count"])-np.log(dictionary["total_data"])
    num_features=len(dictionary[current_class].keys())-1
    for j in range(1,num_features+1):
        xj=x[j-1]
        count_current_class_with_value_xj=dictionary[current_class][j][xj]+1
        count_current_class=dictionary[current_class]["total_count"]+len(dictionary[current_class][j].keys())
        current_xj_probability=(count_current_class_with_value_xj)/(count_current_class)
        output=output*current_xj_probability
    return output
```

```python
def predictSinglePoint(dictionary,x):
    classes=dictionary.keys()
    best_p=-1000
    best_class=-1
    first_run=True
    for current_class in classes:
        probability=Probability(dictionary,x,current_class)
        if(current_class=="total_data"):
            continue
        if(first_run or probability>best_p):
            best_p=probability
            best_class=current_class
        first_run=False
    return best_class
```

```python
def predict(dictionary,X_test):
    Y_pred=[]
    for x in X_test:
        x_class=predictSinglePoint(dictionary,x)
        Y_pred.append(x_class)
    return Y_pred
```

```python
def makeLabelled(column):
    second_limit=column.mean()
    first_limit=0.5*second_limit
    third_limit=1.5*second_limit
    for i in range(0,len(column)):
        if(column[i]<first_limit):
            column[i]=0
        if(column[i]<second_limit):
            column[i]=1
        if(column[i]<third_limit):
            column[i]=2
        else:
            column[i]=3
    return column
```

```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
X=iris.data
```

```python
Y=iris.target
```

```python
for i in range(0,X.shape[-1]):
    X[:,i]=makeLabelled(X[:,i])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
dictionary=fit(X_train,Y_train)
```

```python
Y_pred=predict(dictionary,X_test)
```

# PCA on lfw Images

```python
from sklearn import datasets
```

```python
lfw=datasets.fetch_lfw_people(min_faces_per_person=100,resize=0.4)
```

```python
lfw.keys()
```

```python
lfw.data.shape, lfw.images.shape
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1)
    ax.imshow(lfw.images[i],cmap=plt.cm.bone)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(lfw.data,lfw.target,random_state=0)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
```

```python
pca.fit(x_train)
```

```python
k=0
total=sum(pca.explained_variance_)
currentSum=0
while currentSum/total<0.99:
    currentSum+=pca.explained_variance_[k]
    k=k+1
k
```

```python
pca_lfw=PCA(n_components=k,whiten=True) #whiten
```

```python
x_transformed=pca_lfw.fit_transform(x_train)
```

```python
x_transformed.shape
```

# Reproducing Images

```python
x_approx=pca_lfw.inverse_transform(x_transformed)
```

```python
x_approx=x_approx.reshape((855,50,37))
```

```python
fig=plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1)
    ax.imshow(x_approx[i],cmap=plt.cm.bone)
plt.show()
```

# Eigen faces:

```python
pca_lfw.components_.shape
eigenv=pca_lfw.components_.reshape(k,50,37)
```

```python
fig=plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1)
    ax.imshow(eigenv[i],cmap=plt.cm.bone)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
import time
from sklearn.metrics import confusion_matrix, classification_report
```

```python
rf=RandomForestClassifier()
start=time.time()
rf.fit(x_train,y_train)
end=time.time()
print(end-start)
y_pred=rf.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
x_train_pca=x_transformed
x_test_pca=pca_lfw.transform(x_test)
```

```python
rf2=RandomForestClassifier()
start=time.time()
rf2.fit(x_train_pca,y_train)
end=time.time()
y_pred2=rf2.predict(x_test_pca)
print(confusion_matrix(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

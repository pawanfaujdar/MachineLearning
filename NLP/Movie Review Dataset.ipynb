```python
from nltk.corpus import stopwords
```

```python
stop=set(stopwords.words("english"))
import string
punctuation=list(string.punctuation)
stop.update(punctuation)  #list doesn't have update, set does
stop
```

```python
documents=[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid),category))
```

```python
documents[0:5]
```

```python
import random
random.shuffle(documents)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
def clean_review(words):
    output_words=[]
    for w in words:
        if w.lower() not in stop:
            pos=pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents=[(clean_review(document),category) for document,category in documents] #time consuming for this dataset
```

# Building Feature Set:

```python
training_documents=documents[0:1500]
testing_documents=documents[1500:]
```

```python
all_words=[]
for doc in training_documents:
    all_words+=doc[0]
all_words
```

```python
import nltk
freq=nltk.FreqDist(all_words)
freq
```

```python
common=freq.most_common(15)
common
```

```python
features=[i[0] for i in common]
features
```

```python
def get_feature_dict(words):
    current_features={}
    words_set=set(words)
    for w in features:
        current_features[w]=w in words_set
    return current_features
```

```python
training_data=[(get_feature_dict(doc),category) for doc,category in training_documents]
testing_data=[(get_feature_dict(doc),category) for doc,category in testing_documents]
```

```python
training_data
```

# Naive Bayes:

```python
from nltk import NaiveBayesClassifier
```

```python
classifier=NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)
```

```python
classifier.show_most_informative_features(15)
```

# Count Vectorizer:

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents[0]
```

```python
categories=[category for document,category in documents]
```

```python
text_documents=[" ".join(document) for document,category in documents]
```

```python
text_documents[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(text_documents,categories)
```

```python
count_vec=CountVectorizer(max_features=2000,ngram_range=(1,2))
```

```python
x_train_features=count_vec.fit_transform(x_train)
```

```python
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features=count_vec.transform(x_test)
x_test_features
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(x_train_features,y_train)
```

```python
svc.score(x_test_features,y_test)
```

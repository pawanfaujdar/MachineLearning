```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,random_state=0)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf=MLPClassifier(hidden_layer_sizes=(20,),max_iter=3000)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
clf.coefs_      # coefs== weights
len(clf.coefs_) # no of layers
```

```python
clf.coefs_[0].shape #(no of input units, no of hidden units)
```

```python
clf.coefs_[1].shape #(no of hidden units ,no of output units)
```

```python
clf.intercepts_
len(clf.intercepts_)
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
import pandas as pd
import numpy as np
```

```python
df_train=pd.read_csv("mnist/train.csv")
df_test=pd.read_csv("mnist/test.csv")
```

```python
x_train=df_train.iloc[:,1:]
y_train=df_train.iloc[:,0]
x_test=df_test
```

```python
x_train=x_train.values
y_train=y_train.values
x_test=x_test.values
x_train.shape , x_test.shape, y_train.shape
```

```python
y_train=tf.one_hot(y_train,depth=10)
```

```python
sess=tf.Session()
y_train=sess.run(y_train)
y_train.shape
```

```python
input_width=28
input_height=28
input_channels=1
input_pixels=784

n_conv1=32
n_conv2=64
stride_conv1=1
stride_conv2=1
conv1_k=5
conv2_k=5
max_pool1_k=2
max_pool2_k=2

n_hidden=1024
n_out=10
```

# Initialising weights and biases:

```python
input_size_to_hidden=(input_width//(max_pool1_k*max_pool2_k))*(input_height//(max_pool1_k*max_pool2_k))*(n_conv2)
weights={
    "wc1":tf.Variable(tf.random_normal([conv1_k,conv2_k,input_channels,n_conv1])),
    "wc2":tf.Variable(tf.random_normal([conv2_k,conv2_k,n_conv1,n_conv2])),
    "wh1":tf.Variable(tf.random_normal([input_size_to_hidden,n_hidden])),
    "wo":tf.Variable(tf.random_normal([n_hidden,n_out]))
}

biases={
    "bc1":tf.Variable(tf.random_normal([n_conv1])),
    "bc2":tf.Variable(tf.random_normal([n_conv2])),
    "bh1":tf.Variable(tf.random_normal([n_hidden])),
    "bo":tf.Variable(tf.random_normal([n_out]))
}
```

# Forward Propagation:

```python
def conv(x,weights,bias,strides=1):
    out=tf.nn.conv2d(x,weights,padding="SAME",strides=[1,1,1,1])
    out=tf.nn.bias_add(out,bias)
    out=tf.nn.relu(out)
    return out
```

```python
def maxpooling(x,k=2):
    return tf.nn.max_pool(x,padding="SAME",ksize=[1,k,k,1],strides=[1,k,k,1])
```

```python
def cnn(x,weights,biases,keep_prob):
    x=tf.reshape(x,shape=[-1,input_height,input_width,input_channels])
    conv1=conv(x,weights["wc1"],biases["bc1"],stride_conv1)
    conv1_pool=maxpooling(conv1,max_pool1_k)
    conv2=conv(conv1_pool,weights["wc2"],biases["bc2"],stride_conv2)
    conv2_pool=maxpooling(conv2,max_pool2_k)
    
    hidden_input=tf.reshape(conv2_pool,shape=[-1,input_size_to_hidden])
    hidden_output_before_activation=tf.add(tf.matmul(hidden_input,weights["wh1"]),biases["bh1"])
    hidden_output_before_dropout=tf.nn.relu(hidden_output_before_activation) #*
    hidden_output=tf.nn.relu(hidden_output_before_dropout)
    
    output=tf.add(tf.matmul(hidden_output,weights["wo"]),biases["bo"])
    return output
```

```python
x=tf.placeholder("float",[None,input_pixels])
y=tf.placeholder(tf.int32,[None,n_out])
keep_prob=tf.placeholder("float") #*
pred=cnn(x,weights,biases,keep_prob)
```

```python
cost=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=pred,labels=y))
```

```python
optimizer=tf.train.AdamOptimizer(learning_rate=0.01)
optimize=optimizer.minimize(cost)
```

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
batch_size=100
for i in range(25):
    num_batches=int(x_train.shape[0]/batch_size)
    total_cost=0
    for j in range(num_batches):
        batch_x,batch_y=x_train[j*batch_size:(j+1)*batch_size],y_train[j*batch_size:(j+1)*batch_size]
        c,_=sess.run([cost,optimize],feed_dict={x:batch_x,y:batch_y,keep_prob:0.8})
        total_cost+=c
    print(total_cost) 
```

```python
predictions=tf.argmax(pred,1)
correct_labels=tf.argmax(y,1)
correct_predictions=tf.equal(predictions,correct_labels)
```

```python
predictions,correct_predictions=sess.run([predictions,correct_predictions],feed_dict={x:x_train,y:y_train,keep_prob:1.0})
predictions,correct_predictions
```
